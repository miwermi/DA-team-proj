```python
import pandas as pd
df = pd.read_csv('resources/boardgames2_category_unbucketed.csv',sep=r'\s*,\s*')
df.head()
```

```python
categories = []
columns = ['category1', 'category2', 'category3', 'category4', 'category5', 'category6']
for column in columns:
    for category in df[column]:
        if category not in categories:
            categories.append(category)
categories

```

```python
zeroes = []
for i in range(20000):
    zeroes.append(0)

for category in categories:
    df[category] = zeroes

newdf = df.copy()
newdf.head()
```

```python
for column in columns:
    rowcount = 0
    for item in newdf[column]:
        for category in categories:
            if item == category:
                newdf.at[rowcount,category] = 1
        rowcount+=1
```

```python
newdf = newdf.drop(columns = columns)
```
